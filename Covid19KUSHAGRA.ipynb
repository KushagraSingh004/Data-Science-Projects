```python
#We drop the newcases, newdeaths, and newrecovered columns
df1.drop(["NewCases", "NewDeaths", "NewRecovered"], axis=1, inplace=True)#inplace=True modifies the original dataset
```

```python
px.bar(df1.head(15), x="Country/Region", y="TotalCases", color="TotalCases", hover_data=["Country/Region", "Continent", "WHO Region"], height=500,  color_discrete_sequence=px.colors.sequential.Blues)
```

```python
px.bar(df1.head(15), x="Country/Region", y="TotalCases", color="TotalDeaths", hover_data=["Country/Region", "Continent", "WHO Region", "TotalDeaths"], height=500,  color_discrete_sequence=px.colors.sequential.Blues)
```

```python
df1.head(10)
```

```python
df1["Population"]=df1["Population"].dropna().astype(int)
```

```python
df1.head()
```

```python
df1=df1.dropna(subset=["Population"])
df1["Population"] = df1["Population"].apply(lambda x: int('{:.0f}'.format(float(x))))
```

```python
df1.head()
```

```python
px.bar(df1.head(10), x="Serious,Critical", y="Country/Region", color="ActiveCases", height=500, hover_data=["Population","WHO Region"], orientation='h')
```

```python
px.scatter(df1.head(10), x="Country/Region", y="TotalCases", color="TotalCases", size="TotalCases", size_max=45, hover_data=['Country/Region', 'Continent'], log_y=True)
```
